```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis", model="hipnologo/gpt2-imdb-finetune", tokenizer = "hipnologo/gpt2-imdb-finetune",  truncation=True)
# lvwerra/distilbert-imdb
# distilbert-base-uncased-finetuned-sst-2-english
```

```python
%%time
classifier("I don't enjoy this movie as much as the last, but it was alright, not too bad, not too good, neutral")#[0]["label"]
```

```python
import pandas as pd
df = pd.read_csv("test_data.csv")
```

```python
pred = list()
for index, row in df.iterrows():
  sentiment = (classifier(row["text"]))[0]["label"]
  if sentiment == "LABEL_0":
    pred.append(0)
  else:
    pred.append(1)
```

```python
df["predictions"] = pred
```

```python
sub = pd.DataFrame({
    'Id': df['Id'],
    'Category': df['predictions']
})
sub.to_csv('prediction.csv', index=False)
```
